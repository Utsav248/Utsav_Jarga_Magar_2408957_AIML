### Task 1

````python
def convert_units():
    """
    A program that converts between different units of measurement.
    Supports length (m/ft), weight (kg/lbs), and volume (L/gal) conversions.
    Handles invalid input and conversion types.
    """

    conversion_types = {
        '1': 'Length',
        '2': 'Weight',
        '3': 'Volume'
    }

    length_conversions = {
        '1': ('meters', 'feet', 3.28084),
        '2': ('feet', 'meters', 0.3048)
    }

    weight_conversions = {
        '1': ('kilograms', 'pounds', 2.20462),
        '2': ('pounds', 'kilograms', 0.453592)
    }

    volume_conversions = {
        '1': ('liters', 'gallons', 0.264172),
        '2': ('gallons', 'liters', 3.78541)
    }

    def get_conversion_choice(options, conversion_name):
        """
        Prompts the user to choose a specific conversion within a category.
        Args:
            options (dict): A dictionary of conversion options.
            conversion_name (str): The name of the conversion category (e.g., 'Length').
        Returns:
            str: The user's valid choice.
        """
        print(f"\nChoose {conversion_name} conversion:")
        for key, (from_unit, to_unit, _) in options.items():
            print(f"{key}. Convert {from_unit} to {to_unit}")

        while True:
            choice = input(f"Enter your choice (1-{len(options)}): ")
            if choice in options:
                return choice
            else:
                print("Invalid choice. Please enter a number from the options.")

    def perform_conversion(value, from_unit, to_unit, factor):
        """
        Performs the unit conversion.
        Args:
            value (float): The value to be converted.
            from_unit (str): The unit being converted from.
            to_unit (str): The unit being converted to.
            factor (float): The conversion factor.
        Returns:
            float: The converted value.
        """
        result = value * factor
        print(f"{value} {from_unit} is equal to {result:.2f} {to_unit}")
        return result

    print("Welcome to the Unit Converter!")

    while True:
        print("\nSelect a conversion type:")
        for key, value in conversion_types.items():
            print(f"{key}. {value}")
        print("0. Exit")

        type_choice = input("Enter your choice (1-3, or 0 to exit): ")

        if type_choice == '0':
            print("Exiting Unit Converter. Goodbye!")
            break

        if type_choice not in conversion_types:
            print("Invalid conversion type. Please choose a valid option.")
            continue

        selected_type = conversion_types[type_choice]
        conversion_options = {}
        if selected_type == 'Length':
            conversion_options = length_conversions
        elif selected_type == 'Weight':
            conversion_options = weight_conversions
        elif selected_type == 'Volume':
            conversion_options = volume_conversions

        specific_conversion_choice = get_conversion_choice(conversion_options, selected_type)
        from_unit, to_unit, factor = conversion_options[specific_conversion_choice]

        while True:
            try:
                value_str = input(f"Enter the value in {from_unit} to convert: ")
                value = float(value_str)
                perform_conversion(value, from_unit, to_unit, factor)
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")

convert_units()
````

### Task 2

````python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float or int: The sum of the numbers.
    """
    return sum(numbers)

def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The average of the numbers.

    Raises:
        ValueError: If the list of numbers is empty.
    """
    if not numbers:
        raise ValueError("Cannot calculate average of an empty list.")
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float or int: The maximum number in the list.

    Raises:
        ValueError: If the list of numbers is empty.
    """
    if not numbers:
        raise ValueError("Cannot find maximum of an empty list.")
    return max(numbers)

def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float or int: The minimum number in the list.

    Raises:
        ValueError: If the list of numbers is empty.
    """
    if not numbers:
        raise ValueError("Cannot find minimum of an empty list.")
    return min(numbers)

def perform_math_operations():
    """
    A program that performs various mathematical operations on a list of numbers.
    Supports sum, average, maximum, and minimum operations.
    Handles invalid input and empty lists.
    """

    operations = {
        '1': ('Sum', calculate_sum),
        '2': ('Average', calculate_average),
        '3': ('Maximum', find_maximum),
        '4': ('Minimum', find_minimum)
    }

    print("Welcome to the Math Operations Program!")

    while True:
        print("\nSelect an operation:")
        for key, (name, _) in operations.items():
            print(f"{key}. {name}")
        print("0. Exit")

        choice = input("Enter your choice (1-4, or 0 to exit): ")

        if choice == '0':
            print("Exiting Math Operations Program. Goodbye!")
            break

        if choice not in operations:
            print("Invalid choice. Please select a valid option.")
            continue

        selected_operation_name, selected_function = operations[choice]

        while True:
            numbers_str = input("Enter a list of numbers separated by spaces: ")
            if not numbers_str.strip():
                print("Input cannot be empty. Please enter some numbers.")
                continue

            try:
                # Convert input string to a list of floats
                numbers = [float(x) for x in numbers_str.split()]

                if not numbers:
                    raise ValueError("The list of numbers is empty.")

                result = selected_function(numbers)
                print(f"The {selected_operation_name} of {numbers_str} is: {result}")
                break
            except ValueError as e:
                print(f"Invalid input: {e}. Please enter numeric values.")
            except Exception as e:
                print(f"An unexpected error occurred: {e}")

perform_math_operations()
````

````python
def extract_every_other(lst):
    """
    Extracts every other element from a list, starting from the first element.

    Args:
        lst (list): The input list.

    Returns:
        list: A new list containing every other element from the original list.
    """
    return lst[::2]

# Example usage:
example_list = [1, 2, 3, 4, 5, 6]
result = extract_every_other(example_list)
print(f"Original list: {example_list}")
print(f"Every other element: {result}")
````

### 2. Slice a Sublist

````python
def get_sublist(lst, start, end):
    """
    Returns a sublist from a given list, starting from a specified index
    and ending at another specified index (inclusive).

    Args:
        lst (list): The input list.
        start (int): The starting index.
        end (int): The ending index (inclusive).

    Returns:
        list: The sublist from start to end (inclusive).
    """
    return lst[start:end+1]

# Example usage:
example_list = [1, 2, 3, 4, 5, 6]
start_index = 2
end_index = 4
result = get_sublist(example_list, start_index, end_index)
print(f"Original list: {example_list}")
print(f"Sublist from index {start_index} to {end_index}: {result}")
````

### Task 4: Remove the First and Last Elements

````python
def remove_first_last(lst):
    """
    Removes the first and last elements of a list and returns the resulting sublist.

    Args:
        lst (list): The input list.

    Returns:
        list: A new list without the first and last elements.
    """
    if len(lst) < 2:
        return [] # Returns an empty list if there are fewer than 2 elements
    return lst[1:-1]

# Example usage:
example_list = [1, 2, 3, 4, 5]
result = remove_first_last(example_list)
print(f"Original list: {example_list}")
print(f"List after removing first and last elements: {result}")

example_list_short = [1]
result_short = remove_first_last(example_list_short)
print(f"\nOriginal list (short): {example_list_short}")
print(f"List after removing first and last elements (short): {result_short}")
````

### Task 7: Extract Elements in Reverse Order

````python
def reverse_skip(lst):
    """
    Extracts elements in reverse order starting from the second-to-last
    element and skipping one element in between.

    Args:
        lst (list): The input list.

    Returns:
        list: A new list containing every second element starting from
              the second-to-last, moving backward.
    """
    # Start from the second-to-last element (index -2),
    # go to the beginning (index None),
    # with a step of -2.
    return lst[-2::-2]

# Example usage:
example_list = [1, 2, 3, 4, 5, 6]
result = reverse_skip(example_list)
print(f"Original list: {example_list}")
print(f"Elements in reverse order, skipping one: {result}")
````

## 4.3 Exercise on Nested List

### Task 1: Flatten a Nested List

````python
def flatten(lst):
    """
    Takes a nested list and flattens it into a single list.

    Args:
        lst (list): The nested list.

    Returns:
        list: A flattened version of the list.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(flatten(item)) # Recursively flatten sublists
        else:
            flat_list.append(item)
    return flat_list

# Example usage:
example_list = [[1, 2], [3, 4], [5]]
result = flatten(example_list)
print(f"Original nested list: {example_list}")
print(f"Flattened list: {result}")

example_list_deep = [[1, 2], [3, [4, 5]], 6]
result_deep = flatten(example_list_deep)
print(f"\nOriginal deeply nested list: {example_list_deep}")
print(f"Flattened deeply nested list: {result_deep}")
````

### Task 2: Accessing Nested List Elements

````python
def access_nested_element(lst, indices):
    """
    Extracts a specific element from a nested list given its indices.

    Args:
        lst (list): The nested list.
        indices (list): A list of integers representing the path to the element.

    Returns:
        Any: The element at the specified position.

    Raises:
        IndexError: If any index is out of bounds.
        TypeError: If an intermediate element is not a list when more indices are provided.
    """
    current_element = lst
    for i, index in enumerate(indices):
        if not isinstance(current_element, list):
            raise TypeError(f"Element at path up to index {indices[:i]} is not a list.")
        if index < 0 or index >= len(current_element):
            raise IndexError(f"Index {index} is out of bounds for list of size {len(current_element)} at path {indices[:i]}.")
        current_element = current_element[index]
    return current_element

# Example usage:
example_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(example_list, indices)
print(f"Original nested list: {example_list}")
print(f"Indices to access: {indices}")
print(f"Element at {indices}: {result}")

# Example with deeper nesting
deep_list = [1, [2, [3, 4]], 5]
deep_indices = [1, 1, 0]
deep_result = access_nested_element(deep_list, deep_indices)
print(f"\nOriginal deeply nested list: {deep_list}")
print(f"Indices to access: {deep_indices}")
print(f"Element at {deep_indices}: {deep_result}")

# Example with invalid index
try:
    access_nested_element(example_list, [0, 5])
except (IndexError, TypeError) as e:
    print(f"\nError accessing element: {e}")

# Example with invalid type
try:
    access_nested_element(deep_list, [0, 1])
except (IndexError, TypeError) as e:
    print(f"\nError accessing element: {e}")
````

### Task 3: Sum of All Elements in a Nested List

````python
def sum_nested(lst):
    """
    Calculates the sum of all the numbers in a nested list (regardless of depth).

    Args:
        lst (list): The nested list.

    Returns:
        int or float: The sum of all numeric elements in the list.
    """
    total_sum = 0
    for item in lst:
        if isinstance(item, list):
            total_sum += sum_nested(item)  # Recursively sum sublists
        elif isinstance(item, (int, float)):
            total_sum += item
    return total_sum

# Example usage:
example_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(example_list)
print(f"Original nested list: {example_list}")
print(f"Sum of all elements: {result}")

example_list_2 = [1, [2, [3, [4]]], 5]
result_2 = sum_nested(example_list_2)
print(f"\nOriginal deeply nested list 2: {example_list_2}")
print(f"Sum of all elements 2: {result_2}")
````

### Task 4: Remove Specific Element from a Nested List

````python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Args:
        lst (list): The nested list.
        elem (Any): The element to be removed.

    Returns:
        list: The modified list with all occurrences of elem removed.
    """
    modified_list = []
    for item in lst:
        if isinstance(item, list):
            # Recursively call remove_element for sublists
            modified_list.append(remove_element(item, elem))
        elif item != elem:
            modified_list.append(item)
    return modified_list

# Example usage:
example_list = [[1, 2], [3, 2], [4, 5]]
element_to_remove = 2
result = remove_element(example_list, element_to_remove)
print(f"Original nested list: {example_list}")
print(f"Element to remove: {element_to_remove}")
print(f"List after removing {element_to_remove}: {result}")

example_list_deep = [1, [2, [3, 2, 4]], 5, 2]
element_to_remove_deep = 2
result_deep = remove_element(example_list_deep, element_to_remove_deep)
print(f"\nOriginal deeply nested list: {example_list_deep}")
print(f"Element to remove: {element_to_remove_deep}")
print(f"List after removing {element_to_remove_deep}: {result_deep}")
````

### Task 5: Find the Maximum Element in a Nested List

````python
def find_max(lst):
    """
    Finds the maximum element in a nested list (regardless of depth).

    Args:
        lst (list): The nested list.

    Returns:
        int or float: The maximum element.
    """
    max_val = float('-inf') # Initialize with negative infinity
    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item)) # Recursively find max in sublists
        elif isinstance(item, (int, float)):
            max_val = max(max_val, item)
    return max_val

# Example usage:
example_list = [[1, 2], [3, [4, 5]], 6]
result = find_max(example_list)
print(f"Original nested list: {example_list}")
print(f"Maximum element: {result}")

example_list_2 = [1, [2, [3, [10]]], 5]
result_2 = find_max(example_list_2)
print(f"\nOriginal deeply nested list 2: {example_list_2}")
print(f"Maximum element 2: {result_2}")
````

### Task 6: Count Occurrences of an Element in a Nested List

````python
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Args:
        lst (list): The nested list.
        elem (Any): The element to count.

    Returns:
        int: The number of occurrences of the element.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)  # Recursively count in sublists
        elif item == elem:
            count += 1
    return count

# Example usage:
example_list = [[1, 2], [2, 3], [2, 4]]
element_to_count = 2
result = count_occurrences(example_list, element_to_count)
print(f"Original nested list: {example_list}")
print(f"Element to count: {element_to_count}")
print(f"Number of occurrences: {result}")

example_list_deep = [1, [2, [3, 2, 4]], 5, 2, 2]
element_to_count_deep = 2
result_deep = count_occurrences(example_list_deep, element_to_count_deep)
print(f"\nOriginal deeply nested list: {example_list_deep}")
print(f"Element to count: {element_to_count_deep}")
print(f"Number of occurrences: {result_deep}")
````

### Task 7: Flatten a List of Lists of Lists

````python
def deep_flatten(lst):
    """
    Flattens a list of lists of lists into a single list, regardless of the depth.

    Args:
        lst (list): The deeply nested list.

    Returns:
        list: A single flattened list.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item)) # Recursively flatten sublists
        else:
            flat_list.append(item)
    return flat_list

# Example usage:
example_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(example_list)
print(f"Original deeply nested list: {example_list}")
print(f"Flattened list: {result}")

example_list_2 = [1, [2, [3, [4]]], 5, [[[6]]]]
result_2 = deep_flatten(example_list_2)
print(f"\nOriginal deeply nested list 2: {example_list_2}")
print(f"Flattened list 2: {result_2}")
````

### Task 8: Nested List Average

````python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Args:
        lst (list): The nested list.

    Returns:
        float: The average of all numeric elements in the list.

    Raises:
        ValueError: If the nested list contains no numeric elements to average.
    """
    total_sum = 0
    count = 0

    def get_sum_and_count(current_list):
        nonlocal total_sum, count
        for item in current_list:
            if isinstance(item, list):
                get_sum_and_count(item) # Recursively process sublists
            elif isinstance(item, (int, float)):
                total_sum += item
                count += 1

    get_sum_and_count(lst)

    if count == 0:
        raise ValueError("Cannot calculate average of an empty or non-numeric nested list.")

    return total_sum / count

# Example usage:
example_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(example_list)
print(f"Original nested list: {example_list}")
print(f"Average of all elements: {result}")

example_list_deep = [1, [2, [3, [4]]], 5, [[[6]]]]
result_deep = average_nested(example_list_deep)
print(f"\nOriginal deeply nested list 2: {example_list_deep}")
print(f"Average of all elements 2: {result_deep}")

# Example with no numeric elements
try:
    average_nested([['a'], ['b']])
except ValueError as e:
    print(f"\nError: {e}")
````

### Problem - 4: Matrix Operations

### Problem - 4: Matrix Operations

````python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

print("Initial arrays for Problem 4:")
print(f"A:\n{A}")
print(f"B:\n{B}")
````

#### Task 1: Prove A.A−1 = I.

````python
# 1. Prove A.A−1 = I

# Calculate the inverse of A
A_inv = np.linalg.inv(A)
print(f"\nInverse of A (A_inv):\n{A_inv}")

# Calculate A . A_inv
product_A_A_inv = np.dot(A, A_inv)
print(f"\nProduct A . A_inv:\n{product_A_A_inv}")

# Create an identity matrix of the same size as A
identity_matrix = np.eye(A.shape[0])
print(f"\nIdentity matrix (I):\n{identity_matrix}")

# Check if A . A_inv is close to the identity matrix (due to floating point precision)
# np.allclose is used for numerical comparisons involving floats
is_identity = np.allclose(product_A_A_inv, identity_matrix)
print(f"\nIs A . A_inv approximately equal to I? {is_identity}")
````

#### Task 2: Prove AB ≠ BA.

````python
# 2. Prove AB ≠ BA

# Calculate AB
product_AB = np.dot(A, B)
print(f"\nProduct AB:\n{product_AB}")

# Calculate BA
product_BA = np.dot(B, A)
print(f"\nProduct BA:\n{product_BA}")

# Check if AB is equal to BA
is_equal_AB_BA = np.array_equal(product_AB, product_BA)
print(f"\nIs AB equal to BA? {is_equal_AB_BA}")
print("This demonstrates that matrix multiplication is generally not commutative.")
````

#### Task 3: Prove (AB)T = BTAT

````python
# 3. Prove (AB)T = BTAT

# Calculate AB
product_AB = np.dot(A, B)
print(f"\nProduct AB:\n{product_AB}")

# Calculate (AB)T
AB_transpose = product_AB.T
print(f"\n(AB)T:\n{AB_transpose}")

# Calculate BT
B_transpose = B.T
print(f"\nBT:\n{B_transpose}")

# Calculate AT
A_transpose = A.T
print(f"\nAT:\n{A_transpose}")

# Calculate BTAT
product_B_transpose_A_transpose = np.dot(B_transpose, A_transpose)
print(f"\nBTAT:\n{product_B_transpose_A_transpose}")

# Check if (AB)T is approximately equal to BTAT
is_equal_transpose_product = np.allclose(AB_transpose, product_B_transpose_A_transpose)
print(f"\nIs (AB)T approximately equal to BTAT? {is_equal_transpose_product}")
````

#### Solve the following system of Linear equation using Inverse Methods.

$2x - 3y + z = -1$
$x - y + 2z = -3$
$3x + y - z = 9$

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

````python
import numpy as np

# Represent the system of linear equations in matrix form AX = B
# A is the coefficient matrix
A_system = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

# B is the constants matrix (right-hand side)
B_system = np.array([-1, -3, 9])

print(f"\nCoefficient Matrix A_system:\n{A_system}")
print(f"Constants Matrix B_system: {B_system}")

# Solve for X using the inverse method: X = A_inv * B
# First, calculate the inverse of A_system
A_system_inv = np.linalg.inv(A_system)
print(f"\nInverse of A_system:\n{A_system_inv}")

# Then, multiply A_system_inv by B_system to find the solution vector X
X_solution = np.dot(A_system_inv, B_system)

print(f"\nSolution for x, y, z (X_solution): {X_solution}")

# Verify the solution by plugging X_solution back into the original equations (A_system @ X_solution)
verification = np.dot(A_system, X_solution)
print(f"\nVerification (A_system @ X_solution): {verification}")
print(f"Is verification close to B_system? {np.allclose(verification, B_system)}")
````

## 10.2 Experiment: How Fast is Numpy?

### 1. Element-wise Addition

````python
import numpy as np
import time

SIZE = 1_000_000 # Size of the lists/arrays

# Python Lists - Element-wise Addition
list1 = list(range(SIZE))
list2 = list(range(SIZE))

start_time = time.perf_counter()
result_list = [list1[i] + list2[i] for i in range(SIZE)]
end_time = time.perf_counter()
python_list_time = (end_time - start_time) * 1000 # Convert to milliseconds
print(f"Python Lists - Element-wise Addition: {python_list_time:.4f} ms")

# NumPy Arrays - Element-wise Addition
array1 = np.arange(SIZE)
array2 = np.arange(SIZE)

start_time = time.perf_counter()
result_array = array1 + array2
end_time = time.perf_counter()
numpy_array_time = (end_time - start_time) * 1000 # Convert to milliseconds
print(f"NumPy Arrays - Element-wise Addition: {numpy_array_time:.4f} ms")

print(f"NumPy is approximately {python_list_time / numpy_array_time:.2f} times faster for element-wise addition.")
````

### 4. Matrix Multiplication

````python
import numpy as np
import time

MATRIX_SIZE = 500 # Size of the square matrices (e.g., 500x500)

# Python Lists - Matrix Multiplication (Naive implementation)
def multiply_matrices_py(matrix_a, matrix_b):
    rows_a = len(matrix_a)
    cols_a = len(matrix_a[0])
    rows_b = len(matrix_b)
    cols_b = len(matrix_b[0])

    if cols_a != rows_b:
        raise ValueError("Matrices have incompatible dimensions for multiplication.")

    result_matrix = [[0 for _ in range(cols_b)] for _ in range(rows_a)]

    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):
                result_matrix[i][j] += matrix_a[i][k] * matrix_b[k][j]
    return result_matrix

# Create Python lists representing matrices
matrix_a_py = [[(i * MATRIX_SIZE + j) % 10 for j in range(MATRIX_SIZE)] for i in range(MATRIX_SIZE)]
matrix_b_py = [[(i * MATRIX_SIZE + j) % 10 for j in range(MATRIX_SIZE)] for i in range(MATRIX_SIZE)]

print(f"Performing matrix multiplication for {MATRIX_SIZE}x{MATRIX_SIZE} matrices...")

start_time_py_matrix = time.perf_counter()
result_matrix_py = multiply_matrices_py(matrix_a_py, matrix_b_py)
end_time_py_matrix = time.perf_counter()
python_list_time_matrix = (end_time_py_matrix - start_time_py_matrix) * 1000 # Convert to milliseconds
print(f"Python Lists - Matrix Multiplication: {python_list_time_matrix:.4f} ms")

# NumPy Arrays - Matrix Multiplication
matrix_a_np = np.random.randint(0, 10, size=(MATRIX_SIZE, MATRIX_SIZE))
matrix_b_np = np.random.randint(0, 10, size=(MATRIX_SIZE, MATRIX_SIZE))

start_time_np_matrix = time.perf_counter()
result_matrix_np = np.dot(matrix_a_np, matrix_b_np)
end_time_np_matrix = time.perf_counter()
numpy_array_time_matrix = (end_time_np_matrix - start_time_np_matrix) * 1000 # Convert to milliseconds
print(f"NumPy Arrays - Matrix Multiplication: {numpy_array_time_matrix:.4f} ms")

print(f"NumPy is approximately {python_list_time_matrix / numpy_array_time_matrix:.2f} times faster for matrix multiplication.")
````
